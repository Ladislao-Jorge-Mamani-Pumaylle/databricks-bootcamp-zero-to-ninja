```python
df=spark.read.table("samples.bakehouse.sales_customers")

df_selected = df.select("*")

df_limitado=df_selected.limit(4)
#df_selected.show()
df_limitado.display()
```

## DISTINTC

```python
%sql
SELECT DISTINCT gender FROM samples.bakehouse.sales_customers
```

```python
df_distinct=df_selected.select("country").distinct()
df_distinct.display()
```

```python
df_ciudad = df_selected.groupBy("country") \
                       .count() \
                       #.withColumnRenamed("count", "CANTIDAD_DE_PERSONAS")
df_ciudad.display()
```

```python
df_pivot = df_selected.groupBy("country") \
    .pivot("gender", ["male", "female"]) \
    .count()

df_pivot.display()
```

## WHRERE

```python
%sql
SELECT *
FROM samples.bakehouse.sales_customers
WHERE 1 = 1
AND country="Japan"
AND gender = "female"
LIMIT 5
;
```

```python
from pyspark.sql.functions import col
df_filtered=df_selected.filter(
(col("country") == "USA") & (col("gender") == "female"))

display(df_filtered.limit(5))
```

```python
df_female_japan = df_selected.filter(
    (df_selected.country == "Japan") &
    (df_selected.gender == "female")
).limit(5)

df_female_japan.display()
```
